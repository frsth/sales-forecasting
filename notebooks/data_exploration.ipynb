```python
import pandas as pd
import numpy as np
import datetime


df = pd.read_csv('../data/train_data.csv')
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
# Daily sales total
import seaborn as sns
sns.set_theme()
ax_day = df.groupby('datetime')['Sales'].sum().plot()
ax_day.set_ylabel('Total Sales ($)')
ax_day.set_xlabel('')
ax_day.set_title('Total Sales per Day')
ax_day.set_xlim([datetime.datetime(2014,1,1), datetime.datetime(2016,1,1)])
```

```python
# Daily sales by sub category
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
daily_df = df.groupby(['datetime', 'Sub-Category'])['Sales'].sum().reset_index()
daily_sales_pivot = daily_df.pivot(index='datetime', columns='Sub-Category', values='Sales')

# plot
daily_sales_pivot.iloc[100:200].plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Daily Total Sales by Product')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.legend(title='Product')
plt.tight_layout()
plt.show()
```

```python
# Daily sales by region
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
daily_df = df.groupby(['datetime', 'Region'])['Sales'].sum().reset_index()
daily_sales_pivot = daily_df.pivot(index='datetime', columns='Region', values='Sales')

# plot
daily_sales_pivot.iloc[100:200].plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Daily Total Sales by Region')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.legend(title='Product')
plt.tight_layout()
plt.show()
```
